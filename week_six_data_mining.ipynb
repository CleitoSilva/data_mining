## Using Interquartile Method

```python
Q1 = variables["map_1st"].quantile(0.25)

Q3 = variables["map_1st"].quantile(0.75)

IQR = Q3-Q1

lower_bound=Q1-1.5*IQR

upper_bound=Q3+1.5*IQR

new_cleanData=variables[(variables["map_1st"] >=lower_bound) & (variables["map_1st"] <=upper_bound)]
plt.figure(figsize=(10,8))
sns.boxplot(x=new_cleanData['icu_exp_flg'], y=new_cleanData['map_1st'])
plt.title("Data Amputation (Removing outliers)")
plt.xlabel("ICU Outcome")
plt.ylabel("Mean Arterial Pressure")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt') # download tokenizer data

text="""In Brazil they drive on the right side of the
       road.Brazil as a large coastline on the 
        eastern side of South America."""
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)
```

```python
import docx  

def docxreader(file_path):
    doc = docx.Document(file_path)
    text = ""
    for para in doc.paragraphs:
        text += para.text + " "
    return text

document1 = docxreader("inverted indexes doc 1.docx")
document2 = docxreader("inverted indexes doc 2.docx")

# Tokenize
tokens1 = document1.lower().split()
tokens2 = document2.lower().split()

# Build inverted index
terms = list(set(tokens1 + tokens2))
inverted_index = {}
for term in terms:
    documents = []
    if term in tokens1:
        documents.append("Document 1")
    if term in tokens2:
        documents.append("Document 2")
    inverted_index[term] = documents

# Print inverted index
print("Inverted Index")
for term, documents in inverted_index.items():
    print(f"{term} -> {', '.join(documents)}")
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Count frequencies
word_counts = Counter(tokens1)
common_words = word_counts.most_common(15)

# Plot
words, counts = zip(*common_words)
plt.figure(figsize=(10,6))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 15 Words in Document 1")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


text_all = " ".join(tokens1)

# Create a dark word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",       
    colormap="Set3",            
).generate(text_all)


plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Document 1", color="white") 
plt.show()
```

```python
import PyPDF2
import re
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

pdf_path = 'Machine Learning in Agriculture.pdf'
with open(pdf_path, 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    page_number = 3  
    page_text = pdf_reader.pages[page_number].extract_text()

def process_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\b[a-zA-Z]{1,2}\b', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english')).union({'figure', 'table', 'chapter', 'section', 'et', 'al'})
    filtered_tokens = [word for word in tokens if word not in stop_words]
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(stemmer.stem(word)) for word in filtered_tokens]
    return lemmatized_tokens

def capture_contexts_by_sentence(original_text, keywords):
    sentences = sent_tokenize(original_text)
    keyword_contexts = {}

    for keyword, _ in keywords:
        contexts = []
        for sentence in sentences:
            if re.search(rf'\b{keyword}\b', sentence, re.IGNORECASE):
                clean_sentence = re.sub(r'\s+', ' ', sentence.strip())
                contexts.append(clean_sentence)
        keyword_contexts[keyword] = contexts if contexts else ["No context found"]
    return keyword_contexts


processed_tokens = process_text(page_text)

word_freq = Counter(processed_tokens)
top_words = word_freq.most_common(20)

keyword_contexts = capture_contexts_by_sentence(page_text, top_words)

wordcloud = WordCloud(width=1000,
                      height=600,
                      background_color='black',
                      colormap='plasma',
                      max_words=150).generate(' '.join(processed_tokens))

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Page 4', pad=20)
plt.show()

plt.figure(figsize=(12, 6))
words, counts = zip(*top_words)
plt.barh(words, counts, color='orange')
plt.gca().invert_yaxis()
plt.xlabel('Frequency')
plt.title('Top 20 Keywords on Page 4')
plt.tight_layout()
plt.show()

print("\nTop 20 Keywords on Page 4 with Context:")
print("{:<20} {:<10} {:<}".format('Keyword', 'Count', 'Sample Context'))
print("-" * 90)
for word, count in top_words:
    context_samples = keyword_contexts.get(word, ["No context found"])
    sample_context = context_samples[0]
    print("{:<20} {:<10} {:<}".format(word, count, sample_context))
```

```python
import nltk
import matplotlib.pyplot as plt
import requests
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from wordcloud import WordCloud
from bs4 import BeautifulSoup


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

url="https://www.foxnews.com/?msockid=0e99c386cfa269101e70d62dcedd6804"
response= requests.get(url)
soup=BeautifulSoup(response.text, "html.parser")
text=soup.get_text()

tokens = word_tokenize(text.lower())
stop_words = set(stopwords.words("english"))
filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

stemmer = PorterStemmer()
lemmatizer=WordNetLemmatizer()
stemmed_tokens=[stemmer.stem(word) for word in filtered_tokens]
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]

wordcloud = WordCloud(width=800, height=400, background_color = 'orange', colormap='gray').generate(" ".join(lemmatized_tokens))
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## Principal Component Analysis
### used in dimensionality reduction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer = load_breast_cancer()
X=cancer.data
y=cancer.target
```

```python
X.shape
```

```python
y.shape
```

```python
scaler = StandardScaler()
X_scaled= scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10,8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=50)
plt.title('PCA on breat cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Componen 2')
plt.colorbar(label='Target (0=Malignant, 1=Benign)')
plt.show()
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("The total variance explained by the 2 PCs is:",
     np.sum(pca.explained_variance_ratio_))
```

### Plotting explained variance for all components

```python
pca = PCA()

X_full = pca.fit(X_scaled)

plt.figure(figsize=(10,8))
plt.plot(np.cumsum(X_full.explained_variance_ratio_), marker='o', linestyle='--', color='b')
plt.title('Explained Variance ratio Vs Number of components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

plt.figure(figsize=(10,8))
plt.bar(range(1, len(X_full.explained_variance_ratio_)+1), 
             X_full.explained_variance_ratio_, 
              color='skyblue', edgecolor='k')

plt.title('Individual Explained Variance explained by each principal component')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.show()

```
