## Removing outliers in Data--Data Amputation

### The manual method

```python
cleaned_data = variables[(variables["bmi"]>75)]
plt.figure(figsize=(10,8))
sns.boxplot(x=cleaned_data['icu_exp_flg'], y=cleaned_data['bmi'])
plt.title("Data Amputation (Removing outliers)")
plt.xlabel("ICU Outcome")
plt.ylabel("BMI")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x=data["icu_exp_flg"], y=data["map_1st"])
plt.title("Box Plot of Mean Arterial Pressure in ICU outcome")
plt.xlabel("ICU outcome")
plt.ylabel("Mean Arterial Pressure")
plt.show()
```

```python
cleaned_data = variables[(variables["map_1st"]>=45) & (variables["map_1st"]<=125)]
plt.figure(figsize=(10,8))
sns.boxplot(x=cleaned_data['icu_exp_flg'], y=cleaned_data['map_1st'])
plt.title("Data Amputation (Removing outliers)")
plt.xlabel("ICU Outcome")
plt.ylabel("Mean Arterial Pressure")
plt.show()
```

## Using Interquartile Method

```python
Q1 = variables["map_1st"].quantile(0.25)

Q3 = variables["map_1st"].quantile(0.75)

IQR = Q3-Q1

lower_bound=Q1-1.5*IQR

upper_bound=Q3+1.5*IQR

new_cleanData=variables[(variables["map_1st"] >=lower_bound) & (variables["map_1st"] <=upper_bound)]
plt.figure(figsize=(10,8))
sns.boxplot(x=new_cleanData['icu_exp_flg'], y=new_cleanData['map_1st'])
plt.title("Data Amputation (Removing outliers)")
plt.xlabel("ICU Outcome")
plt.ylabel("Mean Arterial Pressure")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt') # download tokenizer data

text="""In Brazil they drive on the right side of the
       road.Brazil as a large coastline on the 
        eastern side of South America."""
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)
```

```python
import docx  

def docxreader(file_path):
    doc = docx.Document(file_path)
    text = ""
    for para in doc.paragraphs:
        text += para.text + " "
    return text

document1 = docxreader("inverted indexes doc 1.docx")
document2 = docxreader("inverted indexes doc 2.docx")

# Tokenize
tokens1 = document1.lower().split()
tokens2 = document2.lower().split()

# Build inverted index
terms = list(set(tokens1 + tokens2))
inverted_index = {}
for term in terms:
    documents = []
    if term in tokens1:
        documents.append("Document 1")
    if term in tokens2:
        documents.append("Document 2")
    inverted_index[term] = documents

# Print inverted index
print("Inverted Index")
for term, documents in inverted_index.items():
    print(f"{term} -> {', '.join(documents)}")
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Count frequencies
word_counts = Counter(tokens1)
common_words = word_counts.most_common(15)

# Plot
words, counts = zip(*common_words)
plt.figure(figsize=(10,6))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 15 Words in Document 1")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


text_all = " ".join(tokens1)

# Create a dark word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",       
    colormap="Set3",            
).generate(text_all)


plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Document 1", color="white") 
plt.show()
```
