## Data Analysis

```python
import numpy as np # Used for numerical analysis

import pandas as pd # Used for data manipulation

import matplotlib.pyplot as plt # used for data visualisation

import seaborn as sns # used for data visualisation
```

```python
data = pd.read_csv("ICU Dataset.csv")
data
```

```python
variables = data[["age", "bmi", "sofa_first", "weight_first", "icu_exp_flg", "hr_1st" ]]
```

```python
variables
```

### Visualisation of variables of interest using a HeatMap

```python
correlation = variables.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidth=0.5)
plt.title("Correlation of ICU Data")
plt.show()
```

## Box Plots

```python
plt.figure(figsize=(6,6))
sns.boxplot(x=variables["icu_exp_flg"], y=variables["bmi"])
plt.title("Box Plot of Body Mass Index in ICU outcome")
plt.xlabel("ICU outcome")
plt.ylabel("Body Mass index")
plt.show()
```
